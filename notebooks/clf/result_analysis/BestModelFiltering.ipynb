```python
import pandas as pd
from glob import glob 
files = glob("./*")
name = "max_correct"
test_score = "testscores"
results = [i for i in files if name in i and test_score in i]
all_values = pd.concat([pd.read_csv(file, sep="\t") for file in results])
all_values.sort_values(by=['f1_score_test'], ascending=False).head(15)
```

```python
all_values.sort_values(by=['f1_score_test'], ascending=False).to_csv("three_classes_clf_test_scores.csv")
```

```python
sorted_results = all_values.sort_values(by=['f1_score_test'], ascending=False)
```

```python
unique_features = ['clusters', 'AppCats', 'RuLIWC', 'Behavior', 'Words']
```

```python
results_by_unique_features = [(i,[]) for i in sorted_results.columns[:-1]]
for i_feat, feature_name in enumerate(unique_features):
    values = sorted_results[sorted_results["fs"] == feature_name].sort_values(by=['f1_score_test'], ascending=False)[:1].values[0][:-1]
    print(values)
    for i, val in enumerate(values):
        results_by_unique_features[i][-1].append(val)
        
pd.DataFrame(dict(results_by_unique_features))
```
